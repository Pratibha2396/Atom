```python
from IPython.display import clear_output
def display_board(board):
    print('board\n')
    print('   |     |')
    print(' ' + board[7] +  ' | ' +   board[8]  + '   |  ' + board[9])
    print('   |     |')
    print('-------------')
    print('   |     |')
    print(' ' +  board[4] + ' | ' +   board[5] + '   |  ' + board[6])
    print('   |     |')
    print('-------------')
    print('   |     |')
    print(' ' +  board[1] + ' | ' +   board[2] + '   |  ' + board[3])
```

```python
def Player_input():
    
    m=''
    while not(m=='X' or m=='O'):
        m = input('Player 1, choose X or O ').upper()

            
    if m=='O':
        return('O','X')
    else:
        return('X','O')
    
        
```

```python

def place_marker(board, m, position):
    board[position]=m
    display_board(board)
    
```

```python
def win_check(board, m):
    return((board[7]==m and board[8]==m and board[9]==m) or
           (board[4]==m and board[5]==m and board[6]==m) or 
           (board[1]==m and board[2]==m and board[3]==m) or 
           (board[7]==m and board[4]==m and board[1]==m) or 
           (board[8]==m and board[5]==m and board[2]==m) or
           (board[9]==m and board[6]==m and board[3]==m) or 
           (board[1]==m and board[5]==m and board[9]==m) or 
           (board[3]==m and board[5]==m and board[7]==m))
   
    
    
```

```python
import random

def choose_first():
    if random.randint(0, 1) == 0:
        return 'Player 2'
    else:
        return 'Player 1'
```

```python
def space_check(board, position):
    
    return board[position] == ' '
```

```python
def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True
```

```python
def player_choice(board):
    position = 0
    
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input('Choose your next position: (1-9) '))
        
    return position
```

```python
def replay():
    
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
```

```python
print(' Welcome to Tic Tac Toe')

while True:
    gboard= [' ']*10
    p1ch,p2ch=Player_input()
    turn=choose_first()
    print(turn+'will go first')
    
    playgame=input('Are you ready to play?( yes/no)')
    
    if playgame.lower()[0]=='y':
        pgame=True
    else:
        pgame=False
        
    while pgame:
        if turn=='Player 1':
            display_board(gboard)
            pos=player_choice(gboard)
            place_marker(gboard,p1ch,pos)
            
            if win_check(gboard,p1ch):
                display_board(gboard)
                print('YOU won player 1')
                pgame=False
            else:
                if full_board_check(gboard):
                    display_board(gboard)
                    print('TIE!')
                    break
                else:
                    turn='Player 2'
        else :
            display_board(gboard)
            pos=player_choice(gboard)
            place_marker(gboard,p2ch,pos)
            
            if win_check(gboard,p2ch):
                display_board(gboard)
                print('YOU won player 2')
                pgame=False
            else:
                if full_board_check(gboard):
                    display_board(gboard)
                    print('TIE!')
                    break
                else: 
                    turn='Player 1'
    if not replay():
        break
                
        
    
    
```
